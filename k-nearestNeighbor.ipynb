```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('dark')
```

```python
data = pd.read_csv("loan-train.csv")
data.head()
```

```python
data.info()
```

```python
data.drop('Loan_ID', axis='columns', inplace=True)
data.Loan_Status = data.Loan_Status.replace({'Y': 1, 'N' : 0})
data.Gender = data.Gender.replace({'Male': 1, 'Female' : 0})
data.Married = data.Married.replace({"Yes": 1, "No" : 0})
data.Self_Employed = data.Self_Employed.replace({'Yes': 1, 'No' : 0})
```

```python
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)
data['Married'].fillna(data['Married'].mode()[0], inplace=True)
data['Credit_History'].fillna(data['Credit_History'].mean(), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
feature_col = ['Property_Area','Education', 'Dependents']
le = LabelEncoder()
for col in feature_col:
    data[col] = le.fit_transform(data[col])
    data[col] = le.fit_transform(data[col])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']],data.Loan_Status,test_size=0.1)
```

```python
len(X_train)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train_imputed, y_train)
y_pred = knn.predict(X_test_imputed)
accuracy_score(y_test, y_pred)
```
